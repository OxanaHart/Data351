```python
# this aims to strip the start date and end date time that is currently an integer and change it into monthly date ranges as well as create a new column that represents the current month as an integer.

import datetime

import pandas as pd

routes_speed_distance = pd.read_csv("routes_speed_distance.csv")
calendar= pd.read_csv("calendar.txt")
```

```python
calendar['start_date'] = pd.to_datetime(calendar['start_date'].astype(str),format="%Y%m%d")
```

```python
calendar['end_date'] = pd.to_datetime(calendar['end_date'].astype(str),format="%Y%m%d")
```

```python
import datetime

dt_start = calendar.loc[0,'start_date']
dt_end = calendar.loc[0,'end_date']

one_day = datetime.timedelta(1)
start_dates = [dt_start]
end_dates = []
today = dt_start

while today <= dt_end:
    tomorrow = today + one_day
    if tomorrow.month != today.month:
        start_dates.append(tomorrow)
        end_dates.append(today)
    today = tomorrow

end_dates.append(dt_end)

print(start_dates)
print(end_dates)
```

```python
# creating a new data frame with
new_calendar = pd.DataFrame()

for i in range(len(start_dates)):
    new_row = calendar.loc[i]
    new_row['start_date'] = start_dates[i]
    new_row['end_date'] = end_dates[i]
    new_row['month'] = start_dates[i].month
    new_calendar = new_calendar.append(new_row,ignore_index=True)

new_calendar
```
