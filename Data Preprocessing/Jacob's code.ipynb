```python
import pandas as pd
import time



routes = pd.read_csv('routes.txt')
trips = pd.read_csv('trips.txt')
stop_times = pd.read_csv('stop_times.txt',engine="python")
calendar = pd.read_csv('calendar.txt')
calendar_dates = pd.read_csv('calendar_dates.txt')




# take a specific route id
print(routes)
```

```python
specific_route = routes["route_id"][0]

print(specific_route)

```

```python
# match route id to the specific trips

specific_trip = trips.loc[trips["route_id"] == specific_route]

print(specific_trip)
```

```python
# load a specific trip stop times

# this must be done for all trips it must also be check that the trip was the same in the "trip head sign" column

speed_list = []

for i in range(len(specific_trip)):
    trip_id = specific_trip["trip_id"].iloc[i]

    specific_stoptimes = stop_times.loc[stop_times["trip_id"] == trip_id]

    # convert arrival time to timedate format to calculate total time traveled

    specific_stoptimes["arrival_time"] = pd.to_datetime(specific_stoptimes["arrival_time"])
    #start
    start = specific_stoptimes["arrival_time"].iloc[0]
    #finish
    finish = specific_stoptimes["arrival_time"].iloc[-1]
    # time traveled
    total_time = (finish - start).total_seconds()

    # converting time to hours

    hours_traveled = total_time / 60 / 60

    # converting to kilometers
    kms_traveled = specific_stoptimes["shape_dist_traveled"].iloc[-1] /1000

    # calculating speed per hour

    speed = kms_traveled / hours_traveled

    speed_list.append(speed)


print(speed_list)
print(max(speed_list))
print(count(speed_list))
```
